```python
#part 1 :
# Fetch standings, scorers, and recent matches for a league of your
# choice (stick with PL or PD — your pick).
import requests
import pandas as pd
import numpy as np

url_stand = "https://api.football-data.org/v4/competitions/PL/standings"
url_score = "https://api.football-data.org/v4/competitions/PL/scorers"
url_match = "https://api.football-data.org/v4/competitions/PL/matches"

headers = {"X-Auth-Token": "cd007f45d0314e788bc11a08847fcc9c"}

response_stand = requests.get(url_stand,headers=headers ,params={"season": "2025"} )
data_stand = response_stand.json()
standings = data_stand["standings"][0]["table"]
df_stand = pd.json_normalize(standings)

response_score = requests.get(url_score,headers=headers ,params={"season": "2025"} )
data_score = response_score.json()
scorers = data_score["scorers"]
df_score = pd.json_normalize(scorers)


response_match = requests.get(url_match,headers=headers ,params={"season": "2025"} )
data_match = response_match.json()
matches = data_match["matches"]
df_match = pd.json_normalize(matches)

print(df_match)
```
